```python
#import required modules
import numpy as np
import pandas as pd

#Import scraped data
main = pd.read_csv('./BRL_spot_FX_rate_ig.csv')
curr_acct = pd.read_csv('./Current_account_ig.csv')

main.Input_Date = pd.to_datetime(main.Input_Date).dt.date
curr_acct.Input_Date = pd.to_datetime(curr_acct.Input_Date).dt.date
```

```python
#Create Curr_acct_lkup with date ranges
curr_acct.index = curr_acct.Input_Date
curr_acct.head()
```

```python
curr_acct.sort_index(inplace=True)
curr_acct['Start_Date'] = curr_acct.index
curr_acct.head()
```

```python
curr_acct = curr_acct.drop(columns=['Input_Date'])
curr_acct = curr_acct.shift(1)
curr_acct.head()
```

```python
curr_acct['End_Date'] = curr_acct.index
curr_acct.reset_index(inplace=True)
curr_acct.head()
```

```python
main = main.assign(key=1)
curr_acct = curr_acct.assign(key=1)
main.head()
```

```python
main2 = pd.merge(main, curr_acct, on='key').drop('key', axis=1)
main2.head()
```

```python
main2 = main2.query('Input_Date_x >= Start_Date and Input_Date_x < End_Date')
main2.head()
```

```python
main2['Input_Date'] = main2['Input_Date_x']
main2.drop(['Input_Date_x', 'Input_Date_y', 'End_Date', 'Start_Date'], axis=1, inplace=True)
main = main2
main.head(100)
```

```python

Current_account_lag5 = main.Current_account.shift(5)
Current_account_lag5.name = 'Current_account_lag5'
Current_account_lag21 = main.Current_account.shift(21)
Current_account_lag21.name = 'Current_account_lag21'

main = pd.concat([main, Current_account_lag5, Current_account_lag21], axis=1) #.dropna() to drop NA values if needed
main.reset_index(inplace=True)
main.head(60)
```

```python
#Create Curr_acct_lkup with date ranges
curr_acct.index = curr_acct.Input_Date
curr_acct.sort_index(inplace=True)
curr_acct['Start_Date'] = curr_acct.index

curr_acct = curr_acct.drop(columns=['Input_Date'])
curr_acct = curr_acct.shift(1)
curr_acct['End_Date'] = curr_acct.index
curr_acct.reset_index(inplace=True)


#Perform Cartesian join
main = main.assign(key=1)
curr_acct = curr_acct.assign(key=1)

main2 = pd.merge(main, curr_acct, on='key').drop('key', axis=1)
main2 = main2.query('Input_Date_x >= Start_Date and Input_Date_x < End_Date')
main2['Input_Date'] = main2['Input_Date_x']
main2.drop(['Input_Date_x', 'Input_Date_y', 'End_Date', 'Start_Date'], axis=1, inplace=True)
main = main2

Current_account_lag5 = main.Current_account.shift(5)
Current_account_lag5.name = 'Current_account_lag5'
Current_account_lag21 = main.Current_account.shift(21)
Current_account_lag21.name = 'Current_account_lag21'

main = pd.concat([main, Current_account_lag5, Current_account_lag21], axis=1) #.dropna() to drop NA values if needed
main.reset_index(inplace=True)
main.head(60)

```
